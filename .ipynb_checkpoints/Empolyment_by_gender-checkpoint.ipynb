# Empolyment by Gender

```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
jobs_df = pd.read_csv('data/jobsgender.csv')
jobs_df
```

```python
jobs_df.describe()
```

```python
# list(jobs_df.index)
list(jobs_df.columns)
list(jobs_df['minor_category'].unique())
```

```python
cs_df = jobs_df[jobs_df['minor_category']=='Computer and mathematical']
cs_df2 = pd.pivot_table(cs_df, values="percent_female", index=['year'], columns=["occupation"])
cs_df2
```

```python
# cs_df2.plot.scatter(x = index, y = values)
# df.plot(kind='scatter', x='x', y='y', label='Scatter')
cs_df2.plot(figsize=(15,10))
plt.axis([2013,2016,0,100])
plt.title('Percentage of female employees in computer and mathematical occupations')
plt.xlabel('year')
plt.ylabel('female employees percentage')

```

```python
jobs = jobs_df.groupby(['minor_category'])
jobs = jobs['percent_female'].agg([np.mean])
jobs
# jobs.plot(figsize=(15,10))
# plt.title('Percentage of female employees in departments')
# plt.xlabel('job category')
# plt.ylabel('female employees percentage')
# jobs.plot.scatter?
jobs.plot.scatter(jobs['minor_category'], jobs['mean'])
# jobs.columns
```
