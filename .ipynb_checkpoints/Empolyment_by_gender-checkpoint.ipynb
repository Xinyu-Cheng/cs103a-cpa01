# Empolyment by Gender

### This dataset is about women in workplace. It collected the percentage of female workers to male works in different occupations, and also the pay differences. 

This dataset is downloaded from https://www.kaggle.com/kaelihall/women-in-the-workplace?select=jobsgender.csv. To download, you could click the download button on the right. The dataset that I am using is called jobsgender.csv

### Question one: 

How does the percentage of female workers change in the computer and math fields over the years. Does the affirmative action really working?

### Question two:

What category of jobs has more female workers? 

```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
jobs_df = pd.read_csv('data/jobsgender.csv')
jobs_df
```

```python
jobs_df.describe()
```

```python
list(jobs_df.index)
```

```python
list(jobs_df.columns)
```

```python
cs_df = jobs_df[jobs_df['minor_category']=='Computer and mathematical']
cs_df2 = pd.pivot_table(cs_df, values="percent_female", index=['year'], columns=["occupation"])
cs_df2
```

```python
cs_df2.plot(figsize=(15,10))
plt.axis([2013,2016,0,100])
plt.title('Percentage of female employees in computer and mathematical occupations')
plt.xlabel('year')
plt.ylabel('female employees percentage')

```

```python
jobs = jobs_df.groupby(['minor_category'])
jobs = jobs['percent_female'].agg([np.mean])
jobs
```

```python
jobs.plot(figsize=(30,30))
plt.title('Percentage of female employees in departments')
plt.xlabel('job category')
plt.ylabel('female employees percentage')
```
